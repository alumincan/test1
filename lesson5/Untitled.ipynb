```python
cups = int(input("請輸入購買的飲料杯數："))

if cups >= 20:
    discount = '打 7 折'
elif cups >= 10:
    discount = '打 8 折'
elif cups >= 5:
    discount = '打 9 折'
    # 請完成：5~9 杯時,discount 應為「打 9 折」


elif cups >= 1:
    discount = '原價'
    # 請完成：1~4 杯時,discount 應為「原價」


else:
    discount = '沒有購買任何飲料'
    # 請完成：0 杯時,discount 應為「沒有購買任何飲料」


print("折扣方案：", discount)
```
